```python
def sentiment_array(df):
  df_numpy = df.to_numpy()

  X_sentiment = df_numpy[:,-1]

  return X_sentiment.astype(np.float)

X_sentiment = sentiment_array(result_df)
X_sentiment.shape
```

```python
# Creating a numpy array to feed it directly to the network
def windowed_df_to_date_X_y(windowed_dataframe):
  df_as_numpy = windowed_dataframe.to_numpy()

  dates = df_as_numpy[:, 0]

  middle_matrix = df_as_numpy[:,1:-1]
  X = middle_matrix.reshape((len(dates), middle_matrix.shape[1],1))

  Y = df_as_numpy[:,-1]

  return dates, X.astype(np.float32), Y.astype(np.float32)

dates, X, y = windowed_df_to_date_X_y(windowed_df)

dates.shape, X.shape, y.shape
```

```python
q_80 = int(len(dates) * .8)
q_90 = int(len(dates) * .9)

X_sentiment = np.expand_dims(X_sentiment,axis = 1)

X_sentiment_train = X_sentiment[:q_80]
X_sentiment_test = X_sentiment[q_80:q_90]
X_sentiment_val = X_sentiment[q_90:]
```

```python


dates_train, X_train, y_train = dates[:q_80], X[:q_80], y[:q_80]

dates_test, X_test, y_test = dates[q_80:q_90], X[q_80:q_90], y[q_80:q_90]
dates_val, X_val, y_val = dates[q_90:], X[q_90:], y[q_90:]

plt.plot(dates_train, y_train)
plt.plot(dates_val, y_val)
plt.plot(dates_test, y_test)

plt.show
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers, Input
from tensorflow.keras.layers import Reshape, Lambda

# Define input layer for price data
inputs_price = Input(shape=(3, 1))

# Define LSTM layer for price data
x_price = layers.LSTM(64)(inputs_price)

# Define input layer for sentiment model
inputs_sentiment = Input(shape=(1, 1))
inputs_sentiment_scaled = Lambda(lambda x: x * 0.5)(inputs_sentiment) # Multiply by a factor of 2

# Reshape to match price data
inputs_sentiment_reshaped = Reshape((-1,))(inputs_sentiment_scaled)

# Concatenate the outputs of price and sentiment LSTM layers
x = layers.add([x_price, inputs_sentiment_reshaped])


# Define first Dense layer
x = layers.Dense(32, activation='relu')(x)

# Define second Dense layer
x = layers.Dense(32, activation='relu')(x)

# Define output layer
outputs = layers.Dense(1)(x)

# Create the model
model = Model(inputs=[inputs_price, inputs_sentiment], outputs=outputs)

# Compile the model
model.compile(loss='mse', optimizer=Adam(learning_rate=0.00012),
              metrics=['mean_absolute_error'])

# Train the model
model.fit([X_train, X_sentiment_train], y_train, validation_data=([X_val, X_sentiment_val], y_val), epochs=100)


```

```python
train_prediction = model.predict([X_train,X_sentiment_train]).flatten()
test_prediction = model.predict([X_test,X_sentiment_test]).flatten()
val_prediction = model.predict([X_val,X_sentiment_val]).flatten()

plt.plot(dates_train, train_prediction)
plt.plot(dates_train, y_train)
plt.plot(dates_val, val_prediction)
plt.plot(dates_val, y_val)
plt.plot(dates_test, test_prediction)
plt.plot(dates_test, y_test)


plt.legend(['Training Predictions', 'Training Observations',
            'Validation Predictions', 'Validation Observations',
            'Test Predictions', 'Test Observations'])
```

```python
plt.plot(dates_test, test_prediction)
plt.plot(dates_test, y_test)

plt.legend(['Test Predictions', 'Test Observations'])
```

```python
from sklearn import metrics
mae = metrics.mean_absolute_error(y_test, test_prediction)
mse = metrics.mean_squared_error(y_test, test_prediction)
rmse = np.sqrt(mse) # or mse**(0.5)  
r2 = metrics.r2_score(y_test, test_prediction)

print("Results of sklearn.metrics:")
print("MAE:",mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R-Squared:", r2)
```
